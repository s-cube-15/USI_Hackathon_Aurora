```python
def segregate_cols(df):
    num = df.select_dtypes(include = 'number')
    cat = df.select_dtypes(include = 'object')
    return num, cat
```

```python
def scale_data(ip):
    scaler = joblib.load('s_scaler.joblib')
    op = pd.DataFrame(scaler.transform(ip), columns = ip.columns)
    return op
```

```python
def concat_cols(ip1, ip2):
    op = pd.concat([ip1, ip2], axis = 1)
    return op
```

```python
def encode(ip):
    enc = joblib.load('enc.joblib')
    op = pd.DataFrame(enc.transform(ip), columns = ip.columns)
    return op
```

```python
def predict(ip):
    model = joblib.load('model_weights.joblib')
    scaler = joblib.load('s_scaler.joblib')
    enc = joblib.load('enc.joblib')

    ip_num, ip_cat = segregate_cols(ip)
    ip_standard = scale_data(ip_num)
    ip1 = concat_cols(ip_standard, ip_cat)
    ip2 = encode(ip1)
    pred_probab = model.predict_proba(ip2)
    pred = model.predict(ip2)

    if pred == 1:
        pred = 'Not Churn'
    else:
        pred = 'Churn'

    return pred, pred_probab
```

```python
ip = {'rating' : 3, 'feedback_category' : 'Customer Service',	'sentiment' : 'Negative',	'customer_id' : 89669139,	'delivery_status' : 'Slightly Delayed',	'order_total' : 3113.07, 	'payment_method' : 'Wallet',	'total_orders' : 6,	'avg_order_value' : 1999.64}
input = pd.DataFrame([ip])
```

```python
import pandas as pd
import joblib

# Load the model
model = joblib.load('model_weights.joblib')

# Define the required columns
required_columns = ['rating','feedback_category','sentiment','customer_id','delivery_status','order_total','payment_method','total_orders','avg_order_value']

# Load the data
# df = pd.read_excel('sales_data_df.xlsx')
df = pd.read_csv('tatte.csv')

# Ensure the DataFrame contains only the required columns
df_filtered = df[required_columns]

# Initialize lists to store predictions and probabilities
predictions = []
probabilities = []

# Define the predict function
def predict(input_df):
    pred = model.predict(input_df)
    prob = model.predict_proba(input_df)
    return pred, prob

# Process each row
for i, row in df_filtered.iterrows():
    # Convert row to DataFrame (since predict function expects a DataFrame)
    row_df = pd.DataFrame([row.to_dict()])

    # Get prediction and probability
    pred, prob = predict(row_df)

    # Store the results
    predictions.append(pred[0])
    probabilities.append(round(prob[0][1] * 100, 2))

    # Optional: Print progress every 100 rows
    if i % 100 == 0:
        print(f"Processed {i} rows")

# Add the results to the original DataFrame
df['churn_prediction'] = predictions
df['churn_probability'] = probabilities

# Save the updated DataFrame
df.to_csv('data_with_predictions001.csv', index=False)

# Display the first few rows with predictions
print(df[['customer_id', 'churn_prediction', 'churn_probability']].head())
```
